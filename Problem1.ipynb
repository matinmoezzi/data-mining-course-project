### Finding Outliers with Z-Score

```python
from scipy import stats
z = np.abs(stats.zscore(df.iloc[:,2:6]))
z
```

```python
(np.abs(z) > 3)
```

```python
df[(np.abs(z) > 3).any(1)]
```

### Finding Outliers with IQR

```python
Q3 = df.iloc[:,2:6].quantile(0.75)
Q1 = df.iloc[:,2:6].quantile(0.25)
IQR = Q3 - Q1
IQR
```

```python
(df.iloc[:,2:6] < Q1 - 1.5*IQR) | (df.iloc[:,2:6] > 1.5*IQR + Q3)
```

```python
df[((df.iloc[:,2:6] < Q1 - 1.5*IQR) | (df.iloc[:,2:6] > 1.5*IQR + Q3)).any(1)]
```


## Dealing with Outliers

### 1) Logarithmic Transformation

```python
log_df = df.copy()
log_df.insert(4,'Log_Total population',[np.log(x) for x in df['Total population']])
log_df.insert(6,'Log_Area',[np.log(x) for x in df['Area (sq. km)']])
log_df
```

### 2) Marking Outliers

### 2*) Marking based on Z-Score

```python
marked_df = df.copy()
marked_df['IsOutlier'] = np.where((np.abs(z) > 3).any(1), 1, 0)
marked_df
```

### 2**) Marking based on IQR

```python
marked_df['IsOutlier'] = np.where(((df.iloc[:,2:6] < Q1 - 1.5*IQR) | (df.iloc[:,2:6] > 1.5*IQR + Q3)).any(1), 1, 0)
marked_df
```

### 3) Removing Outliers

### 3*) Quantile-based Flooring and Capping

```python
QF = df.iloc[:,2:6].quantile(0.1)
QC = df.iloc[:,2:6].quantile(0.9)
df[~((df.iloc[:,2:6] < QF) | (df.iloc[:,2:6] > QC)).any(1)].shape
```

### 3**) Z-Scored-based Trimming

```python
df[~(np.abs(z) > 3).any(1)].shape
```

# Part (d): Linear Regression

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tools.eval_measures import rmse, mse
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = df.iloc[:,2:5]
X
```

```python
y = df['Coronavirus Cases']
y
```

```python
X = sm.add_constant(X)
linear_model = sm.OLS(y,X).fit()
linear_model.summary()
```

```python
predictions = linear_model.predict(X)
print('RMSE: ', rmse(predictions, y))
print('MSE: ', mse(predictions, y))
```

```python
coeff = pd.DataFrame(data=linear_model.params, columns=['Coefficient'], index=X.columns)
coeff
```

## Part (e): Polynomial Regression

```python
X = df[['Total population']]
poly_features = PolynomialFeatures(degree=2)
poly_X = poly_features.fit_transform(X)
poly_X
```

```python
poly_model = sm.OLS(y, poly_X).fit()
poly_model.summary()
```

```python
poly_coeff = pd.DataFrame(data=poly_model.params, columns=['Coefficient'])
poly_coeff
```

```python
poly_predicts = poly_model.predict(poly_X)
plt.scatter(X, y)
plt.plot(X, poly_predicts)
```
